```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import os,sys,inspect
current_dir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parent_dir = os.path.dirname(current_dir)
sys.path.insert(0, parent_dir) 

import algorithms as a
import force_functions as f
```

```python
K = [1,2,3,4,5]
```

```python
batches = []
for k in K:
    x, _ = a.marsaglia_method(10**k)
    batches.append(x)
```

```python
for batch in batches:
    print(f'batch size: {len(batch)}')
    print(f'mean: {batch.mean()}')
    print(f'std: {batch.std()}')
    print(f'var: {batch.std()**2}\n')
```

```python
def gaussian(x):
    return 1/np.sqrt(2*np.pi) * np.exp(-x**2 / 2.)
```

```python
fig1, ax1 = plt.subplots(figsize=(12,8))
x1 = np.linspace(batches[0].min(), batches[0].max(), 100)
ax1.hist(batches[0], bins=10, density=True)
ax1.plot(x1, gaussian(x1))
ax1.set_xlabel('x')
ax1.set_ylabel('p(x)')
ax1.set_yscale('log')
plt.savefig('./figures/batch10.png', dpi=300, bbox_inches='tight')
plt.show()
```
